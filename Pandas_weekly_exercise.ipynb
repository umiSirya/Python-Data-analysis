# Apple Stock

### Introduction:

We are going to use Apple's stock price.


### Step 1. Import the necessary libraries

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
````

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/09_Time_Series/Apple_Stock/appl_1980_2014.csv)

````python
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/09_Time_Series/Apple_Stock/appl_1980_2014.csv'
````

### Step 3. Assign it to a variable apple

````python
apple = pd.read_csv((url), sep=',')
apple
````

### Step 4.  Check out the type of the columns

````python
apple.info()
````

### Step 5. Transform the Date column as a datetime type

> Add blockquote



````python
apple['Date'] = pd.to_datetime(apple['Date'])
print(apple.dtypes)
print(apple)
````

### Step 6.  Set the date as the index

*   List item
*   List item



````python
apple.set_index('Date', inplace=True)
apple
````

### Step 7.  Is there any duplicate dates?

````python
duplicates = apple.index.duplicated()
apple_duplicates = apple[duplicates] #displays rowws with duplicate dates


print("Duplicates found:", apple_duplicates)

````

### Step 8.  Ops...it seems the index is from the most recent date. Make the first entry the oldest date.

````python
apple.sort_index(inplace=True)
apple
````

### Step 9. Get the last business day of each month

````python
date_range = pd.date_range(start=apple.index.min(), end=apple.index.max(), freq='BM')

last_business_days = pd.DataFrame(index=date_range)

result = last_business_days.merge(apple, left_index=True, right_index=True, how='left')

result
````

### Step 10.  What is the difference in days between the first day and the oldest

````python
oldest_date = apple.index.min()
most_recent_date = apple.index.max()

difference_in_days = (most_recent_date - oldest_date).days
print(f"The difference in days between the oldest and most recent date is: {difference_in_days} days")
````

Step 11.  How many months in the data we have?

````python

unique_months = apple['Date'].dt.to_period('M').nunique()
unique_months
````

### Step 12. Plot the 'Adj Close' value. Set the size of the figure to 13.5 x 9 inches

````python
plt.figure(figsize=(13.5, 9))
plt.plot(apple.index, apple['Adj Close'], marker='o', linestyle='-')
plt.title('Adjusted Close Values')
plt.xlabel('Date')
plt.ylabel('Adjusted Close')
plt.grid()
plt.xticks(rotation=45)
plt.tight_layout()  # Adjust layout to fit labels
plt.show()
````
